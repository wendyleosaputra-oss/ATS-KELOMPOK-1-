Program pada soal nomor 6 menghitung daya listrik dengan rumus:

                                              [P = V \times I]

Tegangan (V) dan arus (I) disimpan sebagai variabel, lalu program menghitung hasilnya dan menampilkan daya listrik (P) dalam satuan Watt.

```python
# Program menghitung daya listrik

# Variabel
V = 220  # tegangan dalam Volt
I = 15   # arus dalam Ampere

# Rumus daya listrik
P = V * I

# Menampilkan hasil
print("Tegangan (V):", V, "Volt")
print("Arus (I):", I, "Ampere")
print("Daya listrik (P):", P, "Watt")
```

Program di atas menghitung efisiensi termal dengan rumus:


                        [Efisiensi = energi keluar​/energi masuk*100]
Pengguna memasukkan dua nilai (energi masuk dan keluar), lalu fungsi menghitung dan menampilkan hasil efisiensinya dalam persen.



```python
# Fungsi untuk menghitung efisiensi termal
def hitung_efisiensi_termal():
    # Meminta input dari pengguna
    energi_masuk = float(input("Masukkan energi masuk (dalam Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (dalam Joule): "))

    # Menghitung efisiensi
    efisiensi = (energi_keluar / energi_masuk) * 100

    # Menampilkan hasil
    print("Efisiensi termal turbin gas adalah:", efisiensi, "%")

# Memanggil fungsi
hitung_efisiensi_termal()
```

Penjelasan Soal No 10

Program di atas mensimulasikan proses pengisian tangki bahan bakar hingga penuh menggunakan perulangan while.

Tangki dimulai dari 0 liter dan kapasitas maksimalnya 100 liter.
Setiap iterasi, tangki bertambah 10 liter, lalu program menampilkan jumlah isi tangki saat ini.

Jika isi tangki mencapai 80 liter, muncul peringatan:

“WARNING : Hampir penuh!”

Ketika tangki sudah mencapai 100 liter, program menampilkan pesan:

“TANGKI PENUH - Pengisian berhenti”

Kemudian perulangan dihentikan dengan perintah break.

Dengan demikian, program ini menggambarkan pengisian bertahap tangki hingga kapasitas maksimum dengan peringatan sebelum penuh dan berhenti otomatis saat tangki sudah penuh.

```python
tangki = 0
kapasitas = 100

while tangki <= kapasitas:
  print(f"tangki berisi: {tangki} liter ")
  if tangki == 80:
    print("WARNING : Hampir penuh! ")
  if tangki == 100:
    print("TANGKI PENUH - Pengisian berhenti")
    break

  tangki += 10
```

Program di atas digunakan untuk menghitung total energi listrik (Wh) yang dihasilkan oleh panel surya selama 7 hari.

Pengguna memasukkan daya panel surya (watt), efisiensi panel, dan lama penyinaran (jam) untuk setiap hari. Program menggunakan fungsi hitung_energi() untuk menghitung energi harian dengan rumus:

Energi

Daya Panel
×
Lama Penyinaran
×
Efisiensi
Energi=Daya Panel×Lama Penyinaran×Efisiensi

Perulangan while digunakan agar perhitungan dilakukan selama 7 hari.
Setiap hasil energi harian dijumlahkan menjadi total energi selama 7 hari.

Setelah perulangan selesai, program menampilkan hasil akhir:

Jika total energi < 2000 Wh → sistem masih kurang efisien.

Jika antara 2000–5000 Wh → kinerja panel cukup baik.

Jika > 5000 Wh → sistem sangat efisien dan optimal.

Terakhir, program menampilkan animasi proses pembuatan laporan menggunakan fungsi time.sleep() untuk memberikan jeda waktu.

Dengan demikian, program ini menggambarkan simulasi efisiensi panel surya berdasarkan input pengguna dengan penerapan fungsi, percabangan if-elif-else, dan perulangan while dalam Python.

```python
import time
def hitung_energi(daya_panel,lama_penyinaran,efisiensi):
  """
  Menghitung energi listrik (Wh) yang di hasilkan panel surya.
  Rumus: Energi = Daya Panel * Lama Penyinaran * Efesiensi
  """
  energi = daya_panel * lama_penyinaran * efisiensi
  return energi
def kategori_penyinaran(jam):
  if jam < 2:
    return "Mendung"
  elif jam <5:
    return "Cerah sebagian"
  else:
    return "Sangat cerah"
print("=== SIMULASI ENERGI SURYA ===")
print("Program ini menghitung total energi listrik dari panel surya dalam 7 hari.\n ")

daya_panel = float(input("Masukkan daya panel surya (watt): "))
efisiensi = float(input("masukan efisiensi panel (contoh 0.8 unntuk 80%)"))
total_energi = 0
hari = 1
while hari <= 7:
  print(f"\nHari ke-{hari}---")
  lama_penyinaran = float(input("Masukkan lama penyinaran (jam): "))
  energi_harian = hitung_energi(daya_panel,lama_penyinaran,efisiensi)
  total_energi += energi_harian
  hari += 1 # Increment hari to avoid infinite loop
print("\n=== HASIL AKHIR ===")
print(f"Total energi listrik selama 7 hari:{total_energi:.2f} Wh")
if total_energi < 2000:
  print("Efisiensi sistem masih rendah , pertimbangkan untuk menambah panel")
elif total_energi < 5000:
  print("Kinerja panel cukup baik. ")
else:
  print("Sistem panel surya sangat efisiensi dan optimal!")
print("\nMenyiapkan laporan",end= "")
for i in range(5):
  print(".",end="",flush=True)
  time.sleep(0.3)
print("\nLaporan energi selesai di buat!")
```
